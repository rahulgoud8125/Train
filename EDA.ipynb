```python
import pandas as pd
import numpy as np
```

```python
d=pd.read_csv('train.csv')
d
```

```python
d.isnull().sum()
```

```python
d=d.drop(columns=['Cabin'])
```

```python
d['Embarked']=d['Embarked'].fillna(d['Embarked'].mode()[0])
```

```python
d['Age']=d['Age'].fillna(d['Age'].median())
```

```python
d.isnull().sum()
```

```python
d['Sex']=d['Sex'].map({'male':0, 'female':1})
d['Embarked']=d['Embarked'].astype('category').cat.codes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=d, x='Survived')
```

```python
What it shows:

-->Number of people who survived (1) vs not survived (0).

-->Gives an idea of class imbalance (more died than survived).

Insight:

-->Majority of passengers did not survive.
```

```python
sns.countplot(data=d, x='Survived', hue='Sex')
```

```python
What it shows:

-->Compare survival between males vs females.

-->Two bars for each group.

Insight:

-->Females survived much more than males.

-->This is because of the policy â€œwomen and children firstâ€.
```

```python
sns.countplot(data=d, x='Pclass', hue='Survived')
```

```python
What it shows:

-->Survival count for all three passenger classes (1st, 2nd, 3rd).

-->Helps compare safety between classes.

Insight:

-->1st class passengers survived the most.

-->3rd class had the lowest survival.
```

```python
sns.histplot(d['Age'], kde=True,bins=10)
```

```python
What it shows:

-->Distribution of passenger ages.

-->KDE line shows smooth curve of density.

Insight:

-->Most passengers were between 20â€“40 years old.

-->Some children and older adults were present.
```

```python
sns.boxplot(data=d, x='Survived', y='Age')
```

```python
What it shows:

-->Age distribution of people who survived vs who did not.

-->Shows median, quartiles, outliers.

Insight:

-->Survivors were slightly younger on average.

-->Many older passengers did not survive.
```

```python
sns.boxplot(data=d, x='Pclass', y='Fare')
```

```python
What it shows:

-->How ticket fare varies in each class.

Insight:

-->1st class paid much higher fare.

-->Shows wide range of fares in upper class.
```

```python
sns.scatterplot(data=d,x='Age',y='Fare',hue='Survived')
```

```python
What it shows:

-->Relationship between age and fare.

-->Shows which points survived (color-coded).

Insight:

-->High-fare passengers (mostly 1st class) survived more.

-->Younger passengers had slightly better survival.
```

```python
sns.heatmap(d[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr(), annot=True,cmap='BuPu')
```

```python
What it shows:

-->Numerical correlation between features.

-->Values between -1 and +1.

Insight:

-->Survival strongly correlates with:

-->Fare (+)

-->Pclass (â€“)

-->Pclass and Fare have strong negative correlation (expected).
```

```python
sns.pairplot(d[['Survived','Pclass','Age','SibSp','Parch','Fare']],palette='BuPu')
```

```python
What it shows:

-->Scatterplots for every pair of variables.

-->Diagonals show distribution of each variable.

Insight:

-->Helps identify:

-->Patterns

-->Clusters

-->Outliers

-->Shows clear grouping between Fare & Pclass.
```
